### 7A. Write a program to shuffle Deck of cards.

```python
import random

# List holders for card values and suits
cardfaces = []
suits = ["Hearts", "Diamonds", "Clubs", "Spades"]
royals = ["J", "Q", "K", "A"]
deck = []

# Add number cards (2-10) to cardfaces
for i in range(2, 11):
    cardfaces.append(str(i))  # This adds numbers 2-10 as strings

# Add royal faces (J, Q, K, A) to cardfaces
for j in range(4):
    cardfaces.append(royals[j])

# Create the deck by combining faces and suits
for k in range(4):  # Loop through suits
    for l in range(13):  # Loop through faces
        card = cardfaces[l] + " of " + suits[k]
        deck.append(card)

# Shuffle the deck
random.shuffle(deck)

# Display the shuffled deck
for m in range(52):
    print(deck[m])
```

### 7B. Solve traveling salesman problem using artificial intelligence technique.

```python
import random
import numpy as np

# Define the distance matrix
cities = [[0, 29, 20, 21],
          [29, 0, 15, 17],
          [20, 15, 0, 28],
          [21, 17, 28, 0]]

# Number of cities
num_cities = len(cities)

# GA parameters
population_size = 100
generations = 500
mutation_rate = 0.01
elitism = 20

# Create an initial population of random tours
def create_population(size):
    population = []
    for _ in range(size):
        tour = list(np.random.permutation(num_cities))
        population.append(tour)
    return population

# Fitness function (total distance of the tour)
def fitness(tour):
    distance = 0
    for i in range(len(tour)):
        distance += cities[tour[i - 1]][tour[i]]
    return distance

# Selection using tournament selection
def selection(population, scores):
    tournament_size = 5
    tournament = random.sample(list(zip(population, scores)), tournament_size)
    return min(tournament, key=lambda t: t[1])[0]

# Crossover: Order Crossover (OX)
def crossover(parent1, parent2):
    child = [-1] * num_cities
    start, end = sorted([random.randint(0, num_cities - 1) for _ in range(2)])
    child[start:end] = parent1[start:end]

    pointer = end
    for gene in parent2:
        if gene not in child:
            if pointer >= num_cities:
                pointer = 0
            child[pointer] = gene
            pointer += 1
    return child

# Mutation: Swap mutation
def mutate(tour):
    if random.random() < mutation_rate:
        i, j = random.sample(range(num_cities), 2)
        tour[i], tour[j] = tour[j], tour[i]
    return tour

# Evolve population
def evolve_population(population):
    scores = [fitness(tour) for tour in population]
    new_population = [selection(population, scores) for _ in range(elitism)]  # Elitism

    while len(new_population) < population_size:
        parent1 = selection(population, scores)
        parent2 = selection(population, scores)
        child = crossover(parent1, parent2)
        new_population.append(mutate(child))

    return new_population

# Run the Genetic Algorithm
population = create_population(population_size)
for gen in range(generations):
    population = evolve_population(population)
    best_tour = min(population, key=lambda tour: fitness(tour))
    best_distance = fitness(best_tour)
    print(f"Generation {gen + 1}, Best distance: {best_distance}")

print("\nBest tour found:")
print(best_tour)
```
